```python
def PlotDartboard(ax, df, reason, column):
    df_sorted = df[df['Reason'] == reason].sort_values(column, ascending=False)
    if df_sorted.empty:
        return  # skip if no data

    num_points = len(df_sorted)
    angles = np.linspace(0, 2 * np.pi, num_points, endpoint=False)
    radii = df_sorted[column].astype(float).values
    labels = df_sorted["Country"].values

    ax.scatter(angles, radii, c=radii, cmap='viridis', s=100, edgecolors='black', zorder=3)

    for angle, radius, label in zip(angles, radii, labels):
        ax.text(angle, radius + 1.5, label, ha='center', va='center', fontsize=7, rotation=np.degrees(angle), rotation_mode='anchor')

    ax.set_yticklabels([])
    ax.set_xticklabels([])
    ax.set_title(f"{reason}", fontsize=12, va='bottom')
    ax.grid(True)
```

```python
unique_reasons = [r for r in df_final['Reason'].unique() if r != "Own Country"]

n = len(unique_reasons)
cols = 2
rows = (n + cols - 1) // cols  # Round up

fig, axs = plt.subplots(rows, cols, subplot_kw={'polar': True}, figsize=(10, 5 * rows))

# Flatten axs for easier iteration
axs = axs.flatten()

for i, reason in enumerate(unique_reasons):
    PlotDartboard(axs[i], df_final, reason, "Average broadband speed (MBPS)")

# Remove extra empty plots if any
for j in range(i + 1, len(axs)):
    fig.delaxes(axs[j])

plt.tight_layout()
plt.show()
```
