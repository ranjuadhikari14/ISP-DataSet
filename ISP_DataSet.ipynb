```python
highly_developed_df = df[df['Reason'] == 'Highly Developed']
```

```python
print(highly_developed_df['Country'])
```

```python
print(highly_developed_df[['Country', 'Average broadband speed (MBPS)', 'Broadband Cost Per Month ($)','Broadband Cost Per Month (Npr)']])
```

```python
highly_developed_df.describe()
```

```python
# Find the row with the maximum broadband cost
most_highly_developed_row = highly_developed_df.loc[
    highly_developed_df['Broadband Cost Per Month ($)'].idxmax()
]

```

```python
# Extract the country name
most_highly_developed_country = most_highly_developed_row['Country']
print(most_highly_developed_country)
```

```python
# Find the row with the minimum broadband cost
most_highly_developed_row = highly_developed_df.loc[
    highly_developed_df['Broadband Cost Per Month ($)'].idxmin()
]
most_highly_developed_country = most_highly_developed_row['Country']
print(most_highly_developed_country)
```
